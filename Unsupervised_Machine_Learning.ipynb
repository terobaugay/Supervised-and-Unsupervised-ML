# Unsupervised Machine Learning

# Principal Component Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'wine.csv')
data
```

```python
data = data.drop('quality',axis = 1)
```

```python
data
```

```python
sns.scatterplot(x = 'fixed acidity',y = 'residual sugar',data = data)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca_df = pca.fit_transform(data)
```

```python
data['pca_1'] = pca_df[:,0]
```

```python
data['pca_2'] = pca_df[:,1]
```

```python
data
```

```python
sns.scatterplot(x = 'pca_1',y = 'pca_2',data = data)
```

```python
# Implementing K Means Clustering
from sklearn.cluster import KMeans
```

```python
x = range(2,11)
```

```python
wcss = []
```

```python
for clusters in range(2,11):
    kmeans=KMeans(n_clusters=clusters)
    kmeans.fit(data.drop(['pca_1','pca_2'],axis = 1))
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(2,11),wcss, marker='*')
plt.grid()
```

```python
# taking optimum number of cluster is 5 
```

```python
kmeans=KMeans(n_clusters=5)
data['clusters_5']=kmeans.fit_predict(data.drop(['pca_1', 'pca_2'], axis=1))
```

```python
sns.scatterplot(x='pca_1', y='pca_2', data=data, hue='clusters_5', palette='bright')
```

```python
sns.kdeplot(x='alcohol', data=data, hue='clusters_5')
```

```python
data.columns
```

```python
data.groupby('clusters_5')['alcohol'].max()
```

```python
# taking clusters 6 
kmeans=KMeans(n_clusters=6)
data['clusters_6']=kmeans.fit_predict(data.drop(['pca_1', 'pca_2','clusters_5'], axis=1))
sns.scatterplot(x='pca_1', y='pca_2', data=data, palette='cool', hue='clusters_6')
```

```python
sns.kdeplot(x='alcohol', data=data, hue='clusters_6')
```

```python
data.groupby('clusters_6')['alcohol'].agg(['max', 'min','mean'])
```

```python
sns.kdeplot(x='density', hue='clusters_5',data=data)
```

```python
data.groupby('clusters_5')['density'].agg(['min', 'max', 'mean'])
```

```python
sns.scatterplot(x='fixed acidity',y='residual sugar',hue='clusters_6', palette='bright',data=data)
```
