```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'wine.csv')
data
```

```python
sample = data.sample(frac = 1)
sample
```

```python
sample = sample.sample(frac = 0.2)
sample
```

```python
sns.kdeplot(x = 'chlorides',data = sample)
```

```python
for col in sample.columns:
    sns.kdeplot(x = col, data = sample)
    plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca_df = pca.fit_transform(sample)
```

```python
sample['d1']=pca_df[:,0]
sample['d2']=pca_df[:,1]
```

```python
sns.scatterplot(x = 'd1',y = 'd2',data = sample)
```

```python
# Implementing K Means Clustering
from sklearn.cluster import KMeans
x = range(2,11)
wcss = []
for clusters in range(2,11):
    kmeans=KMeans(n_clusters=clusters)
    kmeans.fit(sample.drop(['d1','d2'],axis = 1))
    wcss.append(kmeans.inertia_)
plt.plot(range(2,11),wcss, marker='*')
plt.grid()
```

```python
# N clusters =5 
kmeans=KMeans(n_clusters=5)
sample['clusters_5']=kmeans.fit_predict(sample.drop(['d1', 'd2'], axis=1))
```

```python
sns.scatterplot(x='d1', y='d2', data=sample, hue='clusters_5', palette='bright')
```
